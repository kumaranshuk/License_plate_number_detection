```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import os

# Load the trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/data/runs/detect/train4/weights/best.pt')

# Path to the folder containing images which contains 5 images.
img_folder = '/content/drive/MyDrive/data/LPtest'  # Update this to the correct folder path

def crop_number_plate(img_folder):
  l=[]
  # Iterate through each image in the folder
  for img_name in os.listdir(img_folder):
      img_path = os.path.join(img_folder, img_name)


      image = cv2.imread(img_path)

      # Perform object detection
      results = model.predict(source=image, save=False)

      # Extract bounding boxes and process each detected object
      for result in results:
          for box in result.boxes:
              x1, y1, x2, y2 = map(int, box.xyxy[0])
              cropped_img = image[y1:y2, x1:x2]  # Crop the detected object from the image
              l.append(cropped_img)
              # Display the cropped image
              plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
              plt.title(f'Cropped Image: {img_name}')
              plt.show()
  return l
l = crop_number_plate(img_folder)
```

```python
#FOR TEXT EXTRACTION USING OCR(easy ocr)
!pip install easyocr      # For OCR
!pip install opencv-python-headless
```

```python
import easyocr

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'])
for i in l:
  # Perform OCR on the cropped image
  result = reader.readtext(i)

  # Display the extracted text
  plt.imshow(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))
  plt.title(f'Cropped Image: {img_name}')
  plt.show()
  for (bbox, text, prob) in result:
      print(f"Detected text: {text} (Confidence: {prob:.2f})")
```

```python
# The above Ocr trained to work on english language not arabic language that's why it is not giveing perfectly.

#here i have also checked on some english letter License plate look how it will work.
leng = crop_number_plate('/content/drive/MyDrive/data/LPtest_eng')
import easyocr

# Initialize the EasyOCR reader
reader = easyocr.Reader(['en'])
for i in leng:

  # Perform OCR on the cropped image
  result = reader.readtext(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))
  print(result[-1][-1])
  # Display the extracted text
  plt.imshow(cv2.cvtColor(i, cv2.COLOR_BGR2RGB))
  plt.title(f'Cropped Image: {img_name}')
  plt.show()
  for (bbox, text, prob) in result:
      print(f"Detected text: {text} (Confidence: {prob:.2f})")
```
